```python
class Solution:
    def lengthOfLongestSubstring(self, s):
        return self.lengthOfLongestSubstring_asciiSlidingWindow(s)
    
    def lengthOfLongestSubstring_mySolution(self, s):
        table = []
        leftBound = 0
        rightBound = 0
        cnt = 0
        result = 0
        
        while rightBound < len(s):
            c = s[rightBound]
            if c not in table: # if c existed in table
                table.append(c) # push c into table
                cnt += 1 # count the length of substring
                if cnt > result: # if count larger than result, then cover it
                    result = cnt
            else: # if c does not exist in table
                while s[leftBound] != s[rightBound]: #remove the string of begining to repeating character
                    table.remove(s[leftBound])
                    leftBound += 1
                    cnt -= 1
                leftBound += 1
            rightBound += 1
        return result
    def lengthOfLongestSubstring_slidingWindow(self, s):
        n = len(s)
        l = []
        ans = 0
        i = 0
        j = 0
        while i < n and j < n:
            if s[j] not in l:
                l.append(s[j])
                j += 1
                ans = max(ans, j - i)
            else:
                l.remove(s[i])
                i += 1
        return ans
    def lengthOfLongestSubstring_dictSlidingWindow(self, s):
        n = len(s)
        ans = 0
        hashmap = {}
        j = 0
        i = 0
        while j < n:
            if s[j] in hashmap.keys():
                i = max(hashmap[s[j]], i)
            ans = max(ans, j - i + 1)
            hashmap[s[j]] = j + 1
            j += 1
        return ans
    def lengthOfLongestSubstring_asciiSlidingWindow(self, s):
        n = len(s)
        ans = 0
        index = [0 for i in range(128)]
        j = 0
        i = 0
        while j < n:
            i = max(index[ord(s[j])], i)
            ans = max(ans, j - i + 1)
            index[ord(s[j])] = j + 1
            j += 1
        return ans
```
