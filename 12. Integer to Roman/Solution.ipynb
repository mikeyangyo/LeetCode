```python
class Solution:
    def intToRoman(self, num):
        result = []
        numS = str(num)
        nowDigit = 1
        for i in range(len(numS) - 1, -1, -1):
            string = ""
            digit = int(numS[i])
            if nowDigit == 1:
                if digit == 9:
                    string += roman_number['nine']
                    digit -= 9
                elif digit >= 5:
                    string += roman_number['five']
                    digit -= 5
                elif digit == 4:
                    string += roman_number['four']
                    digit -= 5
                for i in range(digit):
                    string += roman_number['one']
                result.insert(0, string)
                nowDigit *= 10
            elif nowDigit == 10:
                if digit == 9:
                    string += roman_number['ninety']
                    digit -= 9
                elif digit >= 5:
                    string += roman_number['fifty']
                    digit -= 5
                elif digit == 4:
                    string += roman_number['fourty']
                    digit -= 5
                for i in range(digit):
                    string += roman_number['ten']
                result.insert(0, string)
                nowDigit *= 10
            elif nowDigit == 100:
                if digit == 9:
                    string += roman_number['ninehundred']
                    digit -= 9
                elif digit >= 5:
                    string += roman_number['fivehundred']
                    digit -= 5
                elif digit == 4:
                    string += roman_number['fourhundred']
                    digit -= 5
                for i in range(digit):
                    string += roman_number['hundred']
                result.insert(0, string)
                nowDigit *= 10
            elif nowDigit == 1000:
                for i in range(digit):
                    string += roman_number['thousand']
                result.insert(0, string)
                nowDigit *= 10
        return "".join(result)
        return result
```

```python
roman_number = {
    "one": "I",
    "four" : "IV",
    "five" : "V",
    "nine" : "IX",
    "ten" : "X",
    "fourty" : "XL",
    "fifty" : "L",
    "ninety" : "XC",
    "hundred" : "C",
    "fourhundred" : "CD",
    "fivehundred" : "D",
    "ninehundred" : "CM",
    "thousand" : "M"
}
```
