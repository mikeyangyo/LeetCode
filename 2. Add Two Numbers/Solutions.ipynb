```python
class Solution:
    def addTwoNumbers(self, l1, l2):
        return self.addTwoNumbers_solutionOnLeetcode(l1, l2)
    
    def addTwoNumbers_mySolution(self, l1, l2):
        # if one of arguments is empty, then return another one
        if l1 == None:
            return l2
        if l2 == None:
            return l1
        
        resultLN = ListNode(0) # create a ListNode object named resultLN to store answer
        result = resultLN # create a variable to control the resultLN
        num = 0 # store the digit of tens place
        
        while l1 and l2: # iterate until one of arguments is empty
            result.next = ListNode((l1.val + l2.val + num) % 10)
            num = int((l1.val + l2.val + num) / 10)
            result = result.next
            l1 = l1.next
            l2 = l2.next
            
        while l1: # if l1 is not empty
            result.next = ListNode((l1.val + num) % 10)
            num = int((l1.val + num) / 10)
            result = result.next
            l1 = l1.next
        
        while l2: # if l2 is not empty
            result.next = ListNode((l2.val + num) % 10)
            num = int((l2.val + num) / 10)
            result = result.next
            l2 = l2.next
            
        if num == 1: # if last digit need to carry
            result.next = ListNode(1)
            
        return resultLN.next
        
        
    def addTwoNumbers_solutionOnLeetcode(self, l1, l2):
        if l1 is None:
            return l2
        if l2 is None:
            return l1
        
        dummy = ListNode(0) # create a ListNode object named dummy to store answer
        current = dummy # create a variable to control the dummy
        carry = 0 # store the digit of carrying
        
        while l1 != None or l2 != None: # iterate until all elements in arguments
            x = l1.val if l1 != None else 0
            y = l2.val if l2 != None else 0
            sum = carry + x + y
            carry = int(sum / 10)
            current.next = ListNode(sum % 10)
            current = current.next
            if l1 != None:
                l1 = l1.next
            if l2 != None:
                l2 = l2.next
            
        if carry > 0: # if last digit need to carry
            current.next = ListNode(carry)
            
        return dummy.next
```
